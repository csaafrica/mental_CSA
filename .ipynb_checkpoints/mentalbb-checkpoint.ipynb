## Income Distribution

```python
# Income distribution among the individuals
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.histplot(mental_health['Income'], bins=50, kde=True, color='skyblue')
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

## Observation:

## The distribution of income is right-skewed, meaning the majority of individuals earn a relatively low income.

## There are a few high-income earners, but they are outliers.

## This suggests income inequality is present in the population and should be taken into account when analyzing economic burden or access to mental health services.

## Income vs History of Mental Illness

```python
# boxplot showing income associated with history of mental illness

import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='History of Mental Illness', y='Income', data=mental_health)
plt.title('Income vs History of Mental Illness')
plt.xlabel('History of Mental Illness')
plt.ylabel('Income')
plt.tight_layout()
plt.show()
```

## Observation:

Individuals with a history of mental illness tend to have lower median incomes compared to those without.

The income spread is wider for people without mental illness, showing more economic diversity among them.

This might indicate that mental health challenges negatively impact earning potential or career progression.

## Employment Status vs History of Mental Illness

```python

plt.figure(figsize=(12, 6))
sns.countplot(data=mental_health, x='Employment Status', hue='History of Mental Illness', palette='Set2')
plt.title('Employment Status vs History of Mental Illness')
plt.xlabel('Employment Status')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Observation:

Unemployed individuals report significantly more cases of mental illness.

Students and retirees also show higher prevalence, while full-time employed individuals show fewer cases.

This suggests a strong link between employment and mental health, possibly due to financial stress, lack of structure, or social isolation.

## Income by Education level

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=mental_health, x='Education Level', y='Income', palette='coolwarm')
plt.title('Income by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Income')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Observation:

Income increases with higher education levels.

Individuals with Masterâ€™s and Bachelorâ€™s degrees earn substantially more than those with just primary or high school education.

This confirms that education is a key economic driver, and lack of access to education could contribute to both low income and mental health vulnerability.

## Mental Illness by Sleep Patterns

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=mental_health, x='Sleep Patterns', hue='History of Mental Illness', palette='Set1')
plt.title('Mental Illness by Sleep Patterns')
plt.xlabel('Sleep Patterns')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

## Observation:

Those with poor sleep patterns are much more likely to report mental illness.

"Fair" sleepers also show higher counts, while those with "Good" sleep report fewer mental health issues.

This reinforces the evidence that sleep quality is a critical factor in mental health, and could be a useful predictor or intervention point.
