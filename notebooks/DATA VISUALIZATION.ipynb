```python
#how to draw the subplots in one plot
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(14, 10))  # 2 rows, 2 columns

# 1️⃣ Line Plot
x = np.arange(13)
y = np.linspace(90, 1400, 13)
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_ylabel("Number of students CSA")
axs[0, 0].set_xlabel("Applicants")
axs[0, 0].set_title("CSA Line Distribution")

# 2️⃣ Scatter Plot
x = [5, 7, 8, 7, 2, 9, 6, 10, 67, 34]
y = [90, 78, 98, 96, 94, 89, 65, 56, 43, 33]
axs[0, 1].scatter(x, y, color='green')
axs[0, 1].set_ylabel("Number of students CSA")
axs[0, 1].set_xlabel("Applicants")
axs[0, 1].set_title("CSA Scatter Distribution")

# 3️⃣ Histogram
data = np.random.normal(0, 1, 100)
axs[1, 0].hist(data, bins=10, color='orange')
axs[1, 0].set_ylabel("Number of students CSA")
axs[1, 0].set_xlabel("Applicants")
axs[1, 0].set_title("CSA Histogram Distribution")

# 4️⃣ Leave Empty or Add Another Plot
axs[1, 1].axis('off')  # You can also use axs[1, 1] for another plot if needed

plt.tight_layout()
plt.show()
```
