```python
#Homework
#Does the ticket fare(fare) ,passenger class(pclass), family onboard(parch) class affect the survival

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
titanic = sns.load_dataset('titanic')

titanic=pd.read_excel("../data/titanic.xlsx")
titanic
# or pd.read_csv() if you're using a file
#how to hande missing data
print(titanic.isnull().sum())

#Relationship between fare and survival
sns.boxplot(x='survived', y='fare', data=titanic)
plt.title("Fare vs Survival")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Ticket Fare")
plt.show()
#Relationship between pclass andd survival rate
sns.countplot(x='pclass', hue='survived', data=titanic)
plt.title("Passenger Class vs Survival")
plt.xlabel("Passenger Class (1 = Upper, 3 = Lower)")
plt.ylabel("Number of Passengers")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()

#Relationship between parch and survival
sns.barplot(x='parch', y='survived', data=titanic)
plt.title("Number of Parents/Children Aboard vs Survival")
plt.xlabel("Parch (Parents/Children Aboard)")
plt.ylabel("Survival Rate")
plt.show()

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Fare vs Survival
sns.boxplot(x='survived', y='fare', data=titanic, ax=axs[0])
axs[0].set_title("Fare vs Survival")
axs[0].set_xlabel("Survived")
axs[0].set_ylabel("Fare")

# Pclass vs Survival
sns.countplot(x='pclass', hue='survived', data=titanic, ax=axs[1])
axs[1].set_title("Pclass vs Survival")
axs[1].set_xlabel("Pclass")
axs[1].set_ylabel("Count")

# Parch vs Survival
sns.barplot(x='parch', y='survived', data=titanic, ax=axs[2])
axs[2].set_title("Parch vs Survival")
axs[2].set_xlabel("Parents/Children Onboard")
axs[2].set_ylabel("Survival Rate")

plt.tight_layout()
plt.show()

```
